O cГіdigo implementa uma rede neural simples para classificar pacientes com diabetes

```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'

#Importanto um DataSet sobre Diabete onde totas as pacientes sao mulheres

df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.columns
```

```python
#Mudando o nome das colunas
df.columns = [
    'Num Gestasoes',
    'Glicose',
    'Pressao Arterial',
    'Espessura da pele',
    'Insulina',
    'IMC',
    'Funcao Pedigree',
    'Idade',
    'Resultado'
]
```

```python
df.head()
```

```python
#Separando meus dados de treinamento

X = df[['Num Gestasoes', 'Glicose', 'Pressao Arterial', 'Espessura da pele', 'Insulina', 'IMC', 'Funcao Pedigree', 'Idade']]
Y = df['Resultado']
```

```python
from sklearn.model_selection import train_test_split

## Divide os dados em conjuntos de treinamento (80%) e teste (20%)

X_treinamento, X_teste, Y_treinamento, Y_teste = train_test_split(X, Y, test_size=0.2, random_state= 42)
```

```python
from sklearn.preprocessing import StandardScaler

##Normalizando os dados

scaler = StandardScaler()
X_treinamneto = scaler.fit_transform(X_treinamento)
X_teste = scaler.transform(X_teste)
```

```python
X.shape
```

```python
modelo = Sequential()
modelo.add(Dense(32, activation='relu', input_shape=(8,))) #Espera 8 valores de entrada
modelo.add(Dense(16, activation='relu'))
modelo.add(Dense(8, activation='relu'))
modelo.add(Dense(1, activation='sigmoid')) # saГ­da binГЎria (0 ou 1)
```

```python
modelo.compile(
    optimizer='adam',
    loss='binary_crossentropy',    # FunГ§ГЈo para classificaГ§ГЈo binГЎria
    metrics=['accuracy']
)
```

```python
modelo.fit(X_treinamento, Y_treinamento, epochs = 100, batch_size=16)
```

Tinha deixado apenas com uma camda oculta, mais decidi adiconar masi uma na rede e deu uma breve melhorada na acuracci

```python
modelo.evaluate(X_teste, Y_teste) #avaliando desempenho do modelo
```

Agora vou adicionar um DropOut, para ver se melhora ainda mais

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

modelo = Sequential()
modelo.add(Dense(32, activation='relu', input_shape=(8,)))
modelo.add(Dropout(0.3))  # desligar 30% dos neurГґnios dessa camada
modelo.add(Dense(16, activation='relu'))
modelo.add(Dropout(0.3))  # outro dropout
modelo.add(Dense(8, activation='relu'))
modelo.add(Dense(1, activation='sigmoid'))  # camada de saГ­da
```

```python
modelo.compile(
    optimizer='adam',
    loss='binary_crossentropy',    # FunГ§ГЈo para classificaГ§ГЈo binГЎria
    metrics=['accuracy']
)
```

```python
modelo.fit(X_treinamento, Y_treinamento, epochs = 100, batch_size=16)
```

```python
modelo.evaluate(X_teste, Y_teste) #avaliando desempenho do modelo
```

Sem camada extra -> 0.6505	0.7169
Com camada extra -> 0.6583	0.6518
Com camada + Dropout -> 0.6505	0.6717

Acredito que o DropOut funcionou evitanto o Overfitting, por isso a acuraccia abaixou
