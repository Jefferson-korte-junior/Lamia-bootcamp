O código implementa uma rede neural simples para classificar pacientes com diabetes

```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'

#Importanto um DataSet sobre Diabete onde totas as pacientes sao mulheres

df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.columns
```

```python
#Mudando o nome das colunas
df.columns = [
    'Num Gestasoes',
    'Glicose',
    'Pressao Arterial',
    'Espessura da pele',
    'Insulina',
    'IMC',
    'Funcao Pedigree',
    'Idade',
    'Resultado'
]
```

```python
df.head()
```

```python
#Separando meus dados de treinamento

X = df[['Num Gestasoes', 'Glicose', 'Pressao Arterial', 'Espessura da pele', 'Insulina', 'IMC', 'Funcao Pedigree', 'Idade']]
Y = df['Resultado']
```

```python
from sklearn.model_selection import train_test_split

## Divide os dados em conjuntos de treinamento (80%) e teste (20%)

X_treinamento, X_teste, Y_treinamento, Y_teste = train_test_split(X, Y, test_size=0.2, random_state= 42)
```

```python
from sklearn.preprocessing import StandardScaler

##Normalizando os dados

scaler = StandardScaler()
X_treinamneto = scaler.fit_transform(X_treinamento)
X_teste = scaler.transform(X_teste)
```

```python
X.shape
```

```python
modelo = Sequential()
modelo.add(Dense(32, activation='relu', input_shape=(8,))) #Espera 8 valores de entrada
modelo.add(Dense(16, activation='relu'))
modelo.add(Dense(8, activation='relu'))
modelo.add(Dense(1, activation='sigmoid')) # saída binária (0 ou 1)
```

```python
modelo.compile(
    optimizer='adam',
    loss='binary_crossentropy',    # Função para classificação binária
    metrics=['accuracy']
)
```

```python
modelo.fit(X_treinamento, Y_treinamento, epochs = 100, batch_size=16)
```

Tinha deixado apenas com uma camda oculta, mais decidi adiconar masi uma na rede e deu uma breve melhorada na acuracci

```python
modelo.evaluate(X_teste, Y_teste) #avaliando desempenho do modelo
```

Agora vou adicionar um DropOut, para ver se melhora ainda mais

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

modelo = Sequential()
modelo.add(Dense(32, activation='relu', input_shape=(8,)))
modelo.add(Dropout(0.3))  # desligar 30% dos neurônios dessa camada
modelo.add(Dense(16, activation='relu'))
modelo.add(Dropout(0.3))  # outro dropout
modelo.add(Dense(8, activation='relu'))
modelo.add(Dense(1, activation='sigmoid'))  # camada de saída
```

```python
modelo.compile(
    optimizer='adam',
    loss='binary_crossentropy',    # Função para classificação binária
    metrics=['accuracy']
)
```

```python
modelo.fit(X_treinamento, Y_treinamento, epochs = 100, batch_size=16)
```

```python
modelo.evaluate(X_teste, Y_teste) #avaliando desempenho do modelo
```

```python
import pandas as pd
import tensorflow as tf
import numpy as np
import sklearn
import scikeras
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as k
from tensorflow.keras import utils as np_utils
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import backend as k

def criar_rede():
    k.clear_session()

    modelo = Sequential()
    modelo.add(Dense(32, activation='relu', input_shape=(8,)))
    modelo.add(Dropout(0.2))  # Desativa 20% dos neurônios
    modelo.add(Dense(16, activation='relu'))
    modelo.add(Dropout(0.2))  # Outro Dropout de 20%
    modelo.add(Dense(8, activation='relu'))
    modelo.add(Dense(1, activation='sigmoid'))  # Camada de saída para classificação binária

    modelo.compile(optimizer='adam',
                   loss='binary_crossentropy',
                   metrics=['accuracy'])

    return modelo
```

```python
rede_neural = KerasClassifier(model = criar_rede, epochs=250, batch_size = 10)
```

```python
resultados = cross_val_score(estimator=rede_neural, X=X ,y=Y, cv=10, scoring="accuracy")
```

# Fazendo Regressao com os dados de Diabetes



```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import backend as k
from tensorflow.keras.optimizers import Adam

# Carregar dados
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'
colunas = [
    'Num Gestasoes',
    'Glicose',
    'Pressao Arterial',
    'Espessura da pele',
    'Insulina',
    'IMC',
    'Funcao Pedigree',
    'Idade',
    'Resultado'
]
df = pd.read_csv(url, names=colunas)

#Definindo X e Y
y = df['Glicose']
X = df.drop(['Glicose'], axis=1)

# Escalonar
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

#Criando modelo
def criar_rede_regressao():
    k.clear_session()
    modelo = Sequential()
    modelo.add(Dense(32, activation='relu', input_shape=(X.shape[1],)))
    modelo.add(Dense(16, activation='relu'))
    modelo.add(Dense(1, activation='linear'))  # Regressão: ativação linear
    modelo.compile(optimizer=Adam(learning_rate=0.001),
                   loss='mean_squared_error',
                   metrics=['mean_absolute_error'])
    return modelo

# Treinar
modelo = criar_rede_regressao()
modelo.fit(X_treinamento, y_treinamento, epochs=100, batch_size=10, validation_split=0.2)

# Avaliar
loss, erro_medio = modelo.evaluate(X_teste, y_teste)
print(f"Erro médio absoluto (MAE): {erro_medio:.2f}")
```
